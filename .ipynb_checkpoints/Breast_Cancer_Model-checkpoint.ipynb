# Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.metrics import plot_confusion_matrix, f1_score
from sklearn.metrics import confusion_matrix
import scikitplot as skplt
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

# Read the dataset

```python
# Read the data
cancer = pd.read_csv('data.csv')
# Data head - Visualize the first 5 rows of the dataset
cancer.head()
```

```python
# Verify the mising values in each column
print('Missing values : ' , cancer.isnull().sum())
```

```python
# Information of the data
print('Info of the data')
cancer.info()
```

# Data preprocessing

```python
# Drop the id and unnamed column
cancer.drop(['id','Unnamed: 32'],axis=1,inplace=True)
cancer.head()
```

```python
# Change the categorical value of diagnosis to a numerical value
cancer['diagnosis'].replace(['M','B'],[0,1],inplace=True)
cancer.tail(10)
```

```python
# Scale the values in the columns to a range of 0 to 100
from sklearn.preprocessing import MinMaxScaler
scale=MinMaxScaler(feature_range=(0,100))
```

```python
# Scale the unnbalanced data
cancer['area_mean']=scale.fit_transform(cancer['area_mean'].values.reshape(-1,1))
cancer['smoothness_mean']=scale.fit_transform(cancer['smoothness_mean'].values.reshape(-1,1))
cancer['compactness_mean']=scale.fit_transform(cancer['compactness_mean'].values.reshape(-1,1))
cancer['concavity_mean']=scale.fit_transform(cancer['concavity_mean'].values.reshape(-1,1))
cancer['concave points_mean']=scale.fit_transform(cancer['concave points_mean'].values.reshape(-1,1))
cancer['symmetry_mean']=scale.fit_transform(cancer['symmetry_mean'].values.reshape(-1,1))
cancer['fractal_dimension_mean']=scale.fit_transform(cancer['fractal_dimension_mean'].values.reshape(-1,1))
cancer['radius_se']=scale.fit_transform(cancer['radius_se'].values.reshape(-1,1))
cancer['texture_se']=scale.fit_transform(cancer['texture_se'].values.reshape(-1,1))
cancer['smoothness_se']=scale.fit_transform(cancer['smoothness_se'].values.reshape(-1,1))
cancer['compactness_se']=scale.fit_transform(cancer['compactness_se'].values.reshape(-1,1))
cancer['concavity_se']=scale.fit_transform(cancer['concavity_se'].values.reshape(-1,1))
cancer['concave points_se']=scale.fit_transform(cancer['concave points_se'].values.reshape(-1,1))
cancer['symmetry_se']=scale.fit_transform(cancer['symmetry_se'].values.reshape(-1,1))
cancer['fractal_dimension_se']=scale.fit_transform(cancer['fractal_dimension_se'].values.reshape(-1,1))
cancer['area_worst']=scale.fit_transform(cancer['area_worst'].values.reshape(-1,1))
cancer['smoothness_worst']=scale.fit_transform(cancer['smoothness_worst'].values.reshape(-1,1))
cancer['compactness_worst']=scale.fit_transform(cancer['compactness_worst'].values.reshape(-1,1))
cancer['concavity_worst']=scale.fit_transform(cancer['concavity_worst'].values.reshape(-1,1))
cancer['concave points_worst']=scale.fit_transform(cancer['concave points_worst'].values.reshape(-1,1))
cancer['symmetry_worst']=scale.fit_transform(cancer['symmetry_worst'].values.reshape(-1,1))
cancer['fractal_dimension_worst']=scale.fit_transform(cancer['fractal_dimension_worst'].values.reshape(-1,1))
cancer['perimeter_se']=scale.fit_transform(cancer['perimeter_se'].values.reshape(-1,1))
```

```python
# Verify the new data
cancer.head()
```

```python
# Save the processed data
cancer.to_csv('Processed_Data.csv')
```

# Machine Learning Models Implementation

```python
# Create the independent and dependent variables
# Drop the features with less correlation with diagnosis column, such as fractal_dimension_mean,
# texture_se, smoothness_se, symmetry_se and fractal_dimension_se
x=np.array(cancer.drop(['diagnosis','fractal_dimension_mean',
                        'texture_se','smoothness_se','symmetry_se',
                        'fractal_dimension_se'],axis=1))
y=np.array(cancer['diagnosis'])
```

```python
# Split the data in training and testing
# 20 % of the data for testing and 80 % for training
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

# Models

```python
# Models
lr=LogisticRegression() # Logistci Regression
svm=SVC(probability=True) # Support Vector Machine
knn=KNeighborsClassifier(n_neighbors=3) # K Neirest Neighbors
dt=DecisionTreeClassifier(max_depth=6) # Decision Trees
# Store all the models in a list
models=[lr,svm,knn,dt]
text = ['LOGISTIC REGRESSION', 'SVM', 'KNN', 'DECISION TREES']
# Maps
maps=[plt.cm.Reds, plt.cm.Blues, plt.cm.Greens, plt.cm.Oranges]
for i in range(len(models)):
    models[i].fit(x_train,y_train)
    y_pred=models[i].predict(x_test)
    # Plot the confusion matrix for each model
    titles=[('Confusion matrix without normalization',None),('Confusion matrix with normalization','true')]
    for title, normalize in titles:
        disp=plot_confusion_matrix(models[i],x_test,y_test,
                                  display_labels=['Beningn', 'Malign'],
                                  cmap=maps[i],
                                  normalize=normalize)
        disp.ax_.set_title(title + ' of %s' %(text[i].lower()))
    plt.show()
    
    # Calculate confusion matrix for metrics
    cm = confusion_matrix(y_test,y_pred)
    # Values
    tp = cm[0, 0]
    tn = cm[1, 1]
    fn = cm[0, 1]
    fp = cm[1, 0]
    # Metrics
    acc = ((tp + tn) / (tp + tn + fp + fn)) * 100
    sen = (tp / (tp + fn)) * 100
    spe = (tn / ( tn + fp )) * 100
    
    # Plot curve ROC - AUC of KNN
    y_proba=models[i].predict_proba(x_test)
    skplt.metrics.plot_roc_curve(y_test, y_proba)
    plt.title('ROC CURVE MODEL %s' %(text[i]))
    plt.show()

    print('F1 Score of %s is: %s '% (str(text[i].lower()), f1_score(y_test,y_pred,average='micro')))
    print('Accuracy of %s is: %s '% (str(text[i].lower()), acc))
    print('Sensibility of %s is: %s '% (str(text[i].lower()), sen))
    print('Specifity %s is: %s '% (str(text[i].lower()), spe))
```
